```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import requests
import os


def draw_plot():
    # Check if the file exists, if not, download it
    file_name = "epa-sea-level.csv"
    if not os.path.exists(file_name):
        url = "https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv" # Common source for this file
        print(f"Downloading {file_name} from {url}...")
        r = requests.get(url)
        with open(file_name, 'wb') as f:
            f.write(r.content)
        print("Download complete.")

    # Import data
    df = pd.read_csv(file_name)

    # Create scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"])

    # First line of best fit (using all data)
    res = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
    x_pred = pd.Series(range(1880, 2051))
    y_pred = res.slope * x_pred + res.intercept
    plt.plot(x_pred, y_pred, 'r', label="Fit: 1880-2050")

    # Second line of best fit (from year 2000 onward)
    df_recent = df[df["Year"] >= 2000]
    res_recent = linregress(df_recent["Year"], df_recent["CSIRO Adjusted Sea Level"])
    x_recent = pd.Series(range(2000, 2051))
    y_recent = res_recent.slope * x_recent + res_recent.intercept
    plt.plot(x_recent, y_recent, 'g', label="Fit: 2000-2050")

    # Labels and title
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.title("Rise in Sea Level")
    plt.legend()

    # Save and return
    plt.savefig("sea_level_plot.png")
    return plt.gca()
```

```python
draw_plot()
```
